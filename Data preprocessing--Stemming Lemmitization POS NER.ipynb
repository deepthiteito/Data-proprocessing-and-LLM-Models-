```python
# Text preprocessing - Stemming
# Stemming algorithm chops off word endings to reduce a word to its base stem. Faster
# less accurate and may produce non dictionary words

import nltk
from nltk.stem import PorterStemmer
```

Stemming is a natural language processing technique that reduces words to their root or base form, called a 'stem.' This is typically done by chopping off suffixes from words. For example, the words 'running,' 'runs,' and 'ran' would all be reduced to the stem 'run.' It's a heuristic process that is faster but generally less accurate than lemmatization, and the resulting stem may not always be a valid word. Stemming is useful for tasks like information retrieval where you want to treat different grammatical forms of a word as the same.

Porter Stemmer is a stemming algorithm that reduces words to their root or stem form by chopping off prefixes and suffixes. It's a heuristic process that often results in a stem that is not a valid word, but it's effective for information retrieval and text analysis tasks where you need to group together words with similar meanings. It's generally faster but less accurate than lemmatization.

```python
stemmer=PorterStemmer()
```

```python
words=['running','rans','runs','easily','fairly']
```

```python
for w in words:
    print(w,"----",stemmer.stem(w)) # word and their corresponding stem
```

```python
# Text preprocessing - Lemmitization
# Lemmitization reduces words to their dictionary root also called as lemma. More accurate
# context aware and slower

from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
```

Lemmatization is a natural language processing technique that reduces words to their dictionary root, also known as a lemma. Unlike stemming, lemmatization is more accurate, context-aware, and often produces valid words. For example, it would reduce 'running,' 'runs,' and 'ran' to the lemma 'run,' similar to stemming, but it can also differentiate between words that have the same stem but different meanings or parts of speech.

```python
lemmatizer=WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
for w in words:
    print(w,"-----",lemmatizer.lemmatize(w,pos="v")) # word as a verb
```

In the lemmatizer.lemmatize(w, pos="v") function call, pos stands for 'Part of Speech'. The WordNetLemmatizer can take an optional pos argument to specify the expected part of speech of the word. When pos="v" is used, it tells the lemmatizer to treat the word w as a verb. This allows the lemmatizer to produce a more accurate lemma for verbs, as it considers the word's grammatical role.

```python
sentence = "The quick brown fox jumps over the lazy dog."
```

Part of Speech (POS) Tagging is a process of assigning grammatical labels (like noun, verb, adjective, adverb, etc.) to each word in a sentence. This helps in understanding the syntactic structure and meaning of text.

For example, in the existing notebook, after tokenizing the sentence "The quick brown fox jumps over the lazy dog.", the nltk.pos_tag(words) function produced tags such as:

DT - determiner (e.g., 'The')
JJ - adjective (e.g., 'quick', 'lazy')
NN - singular noun (e.g., 'brown', 'fox', 'dog')
VBZ - verb, third person singular present (e.g., 'jumps')
IN - preposition (e.g., 'over')

```python
import nltk
nltk.download('punkt_tab')
```

```python
words=nltk.word_tokenize(sentence)
nltk.pos_tag(words)

# DT- determiner,JJ-adjective, NN-Noun Singular, VBZ - verb , RB - adverb, IN -preposition
```

You're asking about Part-of-Speech (POS) tags! These abbreviations are used in natural language processing to categorize words based on their grammatical role in a sentence. Here's what each of the tags you mentioned stands for:

VBZ: Verb, third person singular present (e.g., 'buys', 'runs', 'jumps')
VBG: Verb, gerund or present participle (e.g., 'buying', 'running', 'jumping')
DT: Determiner (e.g., 'a', 'the', 'this')
NNP: Proper noun, singular (e.g., 'UK', 'Apple', 'Tim Cook')
NN: Noun, singular or mass (e.g., 'startup', 'investment', 'dog')
IN: Preposition or subordinating conjunction (e.g., 'for', 'in', 'over')
$: Dollar sign (used for currency)
CD: Cardinal digit (e.g., '1', 'billion', 'two hundred')
CC: Coordinating conjunction (e.g., 'and', 'but', 'or')
VB: Verb, base form (e.g., 'increase', 'run', 'jump')

```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.india-briefing.com/news/india-smartphone-market-q3-2025-vivo-apple-growth-shipments-40430.html/")
soup = BeautifulSoup(response.text,'lxml')
```

```python
paragraphs = soup.find_all('p')
paragraphs_txt=[p.text for p in paragraphs]
```

```python
import re
pattern=r'[^a-zA-Z0-9\s.]'
paragraphs_txt=re.sub(pattern,"",str(paragraphs_txt))
paragraphs_txt=re.sub(r'[0-9]+',"",paragraphs_txt)
paragraphs_txt=paragraphs_txt.lower()
```

```python
import sys
print(sys.executable)
!{sys.executable} -m pip install spacy
!{sys.executable} -m spacy download en_core_web_sm
```

```python
import spacy
nlp=spacy.load("en_core_web_sm")
```

```python
doc=nlp(paragraphs_txt)
```

```python
filtered_words=[token.text for token in doc if not token.is_stop and token.is_alpha]
```

```python
for word in filtered_words:
    token = nlp(word)[0]   # process each word individually
    print(f"{token.text:15} POS:{token.pos_:6} Lemma:{token.lemma_}")
```

```python
# Named Entity Recognition
for word in filtered_words:
    doc=nlp(word)
    for ent in doc.ents:
        print(f"Text: {ent.text:15} Label:{ent.label_}")
```

#Named Entity Recognition (NER) is a technique in natural language processing that identifies and classifies real-world entities mentioned in text. These entities can include names of people, organizations, locations, dates, monetary values, geopolitical entities (like countries, cities, states), time expressions, products, and events.

These labels are the categories identified by the Named Entity Recognition (NER) model for the entities found in the text. Here's what some of them mean:

CARDINAL: Numerical values that do not refer to a specific time (e.g., 'million', '1 billion').
DATE: Absolute or relative dates or periods (e.g., 'october', 'july', 'annual').
ORG: Companies, agencies, institutions, etc. (e.g., 'omdia', 'samsung').
GPE: Geopolitical entities, i.e., countries, cities, states (e.g., 'india').
